# Data Exploration

## Loading the Data

```python
from IPython.display import display
import ml_utils as mu

class_labels = ["Control", "Parkinsons", "MSA", "PSP"]

data = mu.get_training_data()
X, y = mu.split_x_and_y(data)

display(data.head())
display(data.info())
display(data.describe())
```

## Data Distribution

```python
%matplotlib inline

ax = data.groupby('GroupID').hist(bins=20, figsize=(15,15))
```

There are 4 groups of feature grids above, one for each class ID in the typical order (control, pd, mda, psp).  For the most part these distributions look normally distributed.  Some outliers seen (can look to remove outliers before training in future)

Some features to be cautious of:
GP_FW (heavily skewed in the first 3 groups)
For MSA:  pSN_FW, MCP_FW, RN_FW

For the atypical groups, non-normality in the these distributions is likely just a result of a lack of data

In general, nothing jumps out as needing to be immediately addressed in this data.

## Evaluating Class Balance

```python
%matplotlib inline

class_counts = data['GroupID'].replace({i:class_labels[i] for i in range(len(class_labels))}).value_counts(normalize=True)

ax = class_counts.plot('bar')
_=ax.set_title("Class Frequencies")
_=ax.set_ylabel("Count")

print(class_counts)
```

There is clearly a class imbalance problem.  Will have to use resampling techniques or imbalance-agnostic algorithms

## More Visual Data Exploration 

```python
import pandas as pd
import yellowbrick as yb 
import matplotlib.pyplot as plt
%matplotlib inline

plt.rcParams['figure.figsize'] = (12, 8)

from yellowbrick.features.rankd import Rank2D
from yellowbrick.features.radviz import RadViz
from yellowbrick.features.pcoords import ParallelCoordinates
```

### Covariance Comparison

```python
Xm = X.as_matrix()
ym = y.as_matrix()

# Instantiate the visualizer with the Covariance ranking algorithm 
visualizer = Rank2D(features=X.columns.values, algorithm='covariance')

visualizer.fit(Xm, ym) 
visualizer.transform(Xm)
visualizer.poof()
```

We don't see much in way of covariance between data.  

Age/UPDRS have strong positive covariance
Sex/UPDRS have strong negative covariance

### Pearson Correlation

```python
Xm = X.as_matrix()
ym = y.as_matrix()

# Instantiate the visualizer with the Covariance ranking algorithm 
visualizer = Rank2D(features=X.columns.values, algorithm='pearson')

visualizer.fit(Xm, ym) 
visualizer.transform(Xm)
visualizer.poof()
```

Some interesting patterns to note:
* All the FW and FA values are negatively correlated
* Age, Sex, and UPDRS are positively correlated with FW values and negatively correlated with FA values
